```python
from langchain_experimental.agents import create_csv_agent
from langchain_anthropic import ChatAnthropic
from dotenv import load_dotenv

load_dotenv()

model = ChatAnthropic(model="claude-3-5-sonnet-20241022", temperature=0.0)

agent = create_csv_agent(
    llm=model,
    path="../retail_store_inventory.csv",
    verbose=True,
    allow_dangerous_code=True,
)
agent.run("What is the total number of items in the inventory?")
agent.run("What is the total number of items in the inventory? How many of each item are there?")   
```
